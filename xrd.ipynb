```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import *
from matplotlib import pyplot
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
#Structure cristalline 
struct = MPRester("BTfRdPWCgouk2CI3AZ6").get_structure_by_material_id("mp-8882")
```

```python
# Diffraction à rayons X d'un structure cristalline en 
#  imposant la longueur d'onde des rayons X à CuKa de λ = 1.54060 Å
xrd = XRDCalculator(wavelength = 1.54060)

# Calcul du pattern de diffraction pour une structure donnée en imposant l'intensité du plus grand pic à 100
pattern = xrd.get_pattern(struct, scaled=True)

# Obtention des indices hkl 
hkl = pattern.hkls
```

# Difractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å sur le cristal de GaP

```python
# Plot des pics avec 2θ allant de 0 à 90 degrés 
xrd.show_plot(struct, annotate_peaks=False)
```

# Zoom sur les trois premiers pics du diffractogramme 

```python
# Plot des 3 premiers pics avec 2θ allant de 10 à 35 degrés 
xrd.show_plot(struct, two_theta_range=(10,35),annotate_peaks=True,ax=None,with_labels=True)
```

```python
#Affichage des données des trois premiers pics : 

print("Premier pic :")
print("--------------")
print("Angle 2θ (en °) :", pattern.x[0])
print("Intensité du pic (en %) :", pattern.y[0])
print('Indices hkil asoosciés :', hkl[0][0]["hkl"], 'de multiplicité', hkl[0][0]["multiplicity"], 'avec i = -h-k')
print('Distance interréticulaire : ', pattern.d_hkls[0])
print('\n')

print("Second pic : ")
print("--------------")
print("Angle 2θ (en °) :", pattern.x[1])
print("Intensité du pic (en %) :", pattern.y[1])
print('Indices hkil asoosciés :', hkl[1][0]["hkl"], 'de multiplicité', hkl[1][0]["multiplicity"], 'avec i = -h-k')
print('Distance interréticulaire : ', pattern.d_hkls[1])
print('\n')

print("Troisième pic : ")
print("--------------")
print("Angle 2θ (en °) :", pattern.x[2])
print("Intensité du pic (en %) :", pattern.y[2])
print('Indices hkil asoosciés :', hkl[2][0]["hkl"], 'de multiplicité', hkl[2][0]["multiplicity"], 'avec i = -h-k')
print('Distance interréticulaire : ', pattern.d_hkls[2])
print('\n')
```

# Explications 

Chaque diffractogramme sert de signature pour la phase cristalline analysée. 
Les pics de difraction sont observés suite à des interférences constructives entre les faisceaux diffusés par l’échantillon analysé. 
La position des pics permet de déterminer les vecteurs du réseau réciproque et donc de décrire la maille cristalline. 



Les pics sont déterminés grâce à la loi de Bragg :
2dsinθ = nλ avec

d = distance interréticulaire

θ = demi-angle de déviation 

n = ordre de la réflexion <

λ = longueur d'onde des rayons X (ici, λ = 1.54060 Å)



Il est intéressant d'expliquer pourquoi en comparant ce diffractogramme à celui de de Material Projects, seules les intensités des pics sont différentes. Comme nous le savons, l'intensité de chacun des pics est proportionnelle au carré de la valeur absolue du facteur de maille (voir photo). En d'autres mots, les intensités sont définies à un facteur près. Il n'est donc pas étonnant que les intensités correspondent en relatif et pas en absolu.
