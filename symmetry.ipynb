```python
from pymatgen.ext.matproj import MPRester
from jupyter_jsmol.pymatgen import quick_view
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from tabulate import tabulate
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pymatgen.symmetry.site_symmetries import *
from pymatgen.core.operations import *
import matplotlib.image as img
```

```python
struct = MPRester("BTfRdPWCgouk2CI3AZ6").get_structure_by_material_id("mp-8882")
```

```python
Ope_Sym1 = SymmOp.from_xyz_string('-x, -y, z+1/2') #4
Ope_Sym2 = SymmOp.from_xyz_string('x, x-y, z')     #8
Ope_Sym3 = SymmOp.from_xyz_string('y, x, z+1/2')   #7

#Différents atomes choisis :
P_1  = [0.33333333, 0.66666667, 0.37406100]
Ga_2 = [0.66666667, 0.33333333, 0.49994]
Ga_3 = [0.33333333, 0.66666667, 0.99993900]

sym1 = Ope_Sym1.operate(P_1)
sym2 = Ope_Sym2.operate(Ga_2)
sym3 = Ope_Sym3.operate(Ga_3)
```

# Première opération de symétrie :

```python
#Affichage des résultats
print('----------------------------------')
print("Première opération de symétrie (n°4)")
print("(" + str(Ope_Sym1.as_xyz_string()) + ")")
print(Ope_Sym1)
print('----------------------------------\n')
print('----------------------------------')
print("Atome de départ : ")
print(str(P_1))
print('----------------------------------\n')
print('----------------------------------')
print("Atome d'arrivée : ")
print(str(sym1))
print('----------------------------------\n')
print('----------------------------------')
print("Analyse : ")
print("Rotation d'ordre 2 et d'axe c \nTranslation de (0, 0, 0.5) ")
print('----------------------------------\n')
print('Image : ')

#Visualisation : 
view = quick_view(struct)
display(view)
view.script('draw symop 4 {atomno = 2}')

image_1=img.imread('ope1.png')
fig, axes = plt.subplots(figsize=(17,15))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)

image_2=img.imread('ope12.png')
fig, axes = plt.subplots(figsize=(17,15))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)

plt.show()
```
